# Dynamic CAPM Equity Analysis

This notebook mirrors the dashboard workflow: load market data, compute CAPM expected returns, calculate risk metrics, and review portfolio-style outputs for the default ticker set.

```python
from src.data_loader import load_price_data, calculate_daily_returns, fetch_risk_free_rate
from src.capm import build_capm_table, build_projection_table
from src.risk_metrics import annualized_return, build_risk_metrics_table, correlation_matrix
from src.utils import DEFAULT_TICKERS, DEFAULT_BENCHMARK, default_date_range
```

```python
start_date, end_date = default_date_range()
price_data = load_price_data(DEFAULT_TICKERS, DEFAULT_BENCHMARK, start_date, end_date)
prices = price_data.prices
returns = calculate_daily_returns(prices)
stock_returns = returns[price_data.stock_tickers]
benchmark_returns = returns[price_data.benchmark]
```

```python
risk_free = fetch_risk_free_rate()
market_return = annualized_return(benchmark_returns)
capm_results = build_capm_table(stock_returns, benchmark_returns, risk_free.rate, market_return)
risk_metrics = build_risk_metrics_table(stock_returns, benchmark_returns, risk_free.rate)
projections = build_projection_table(prices.iloc[-1], capm_results, horizon_months=12)
capm_results
```
